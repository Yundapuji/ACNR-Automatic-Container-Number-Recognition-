```python
#Import Library
import os
from glob import glob
import pandas as pd
from functools import reduce
from xml.etree import ElementTree as et
```

### Ekstrak Objek Info dari File Annotation (.xml)

```python
#Load semua xml file dan simpan di list
xmlfiles = glob('./Gambar/*.xml') #Path semua file xml
```

```python
xmlfiles #Tidak Wajib di run ulang
```

```python
#Data cleaning untuk mengganti '\\' ke '/'
xmlfiles = list(map(lambda x : x.replace('\\' , '/'),xmlfiles))
```

```python
xmlfiles #Tidak Wajib di run ulang
```

```python
#Membaca file xml dan ekstrak informasi
#----1. Read xml file
#----2. Dari file xml ekstrak informasi (filename, size(width, height), object(name, xmin, xmax, ymin, ymax))
def ExTeks(filename):
    tree = et.parse(filename)
    root = tree.getroot()

    #***********Ekstrak filename
    image_name = root.find('filename').text
    #Menghapus spasi pada nama file
    for i in image_name:
        image_name = image_name.replace(' ', '') 
    
    #***********Ekstrak width dan height
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text

    #***********Ekstrak object
    objs = root.findall('object')
    parser = []
    for obj in objs:
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name,width,height,name,xmin,xmax,ymin,ymax])
    
    return parser
```

```python
parser_all = list(map(ExTeks,xmlfiles))
```

```python
parser_all #Tidak Wajib di run ulang
```

```python
#Menyederhanakan multi-dimensional array
data = reduce(lambda x,y : x+y, parser_all)
```

```python
data #Tidak Wajib di run ulang
```

```python
df = pd.DataFrame(data, columns = ['filename', 'width', 'height', 'name', 'xmin', 'xmax', 'ymin', 'ymax'])
```

### Menyiapkan Label Untuk Yolo Model

```python
#Konversi tipe data koordinat bounding box dan size gambar
cols = ['width','height','xmin', 'xmax', 'ymin', 'ymax']
df[cols] = df[cols].astype(int)
df.info()
```

```python
#Menerapkan conversion formula untuk mendapatkan yolo label
#***********Center X, Center Y
df['center_x'] = ((df['xmax'] + df['xmin'])/2)/df['width']
df['center_y'] = ((df['ymax'] + df['ymin'])/2)/df['height']

#***********Width
df['w'] = (df['xmax'] - df['xmin'])/df['width']

#***********Height
df['h'] = (df['ymax'] - df['ymin'])/df['height']
```

### Membagi Data Training Dan Testing (80:20)

```python
#Mengambil nama gambar
images = df['filename'].unique() 
len(images)
```

```python
#Membagi data training (80%) dan data testing (20%)
img_df = pd.DataFrame(images,columns = ['filename'])
img_train = tuple(img_df.sample(frac = 0.8)['filename']) #Shuffle dan ambil 80%
len(img_train)
```

```python
#Warning : Whitespace di awal nama file tidak beraturan
img_train #Tidak Wajib di run ulang
```

```python
img_test = tuple(img_df.query(f'filename not in {img_train}')['filename']) #Shuffle dan ambil 20% gambar sisanya
len(img_test)
```

```python
#Warning : Whitespace di awal nama file tidak beraturan
img_test #Tidak Wajib di run ulang
```

```python
train_df = df.query(f'filename in {img_train}')
test_df = df.query(f'filename in {img_test}')
```

### Label Encoding

```python
def label_encoding (x):
    labels = {'container-number':0}
    return labels[x]
```

```python
train_df['id'] = train_df['name'].apply(label_encoding)
test_df['id'] = test_df['name'].apply(label_encoding)
```

### Save Image Dan Label Dalam Bentuk .txt

```python
#Import Library
import os
from shutil import move
```

```python
train_folder = 'Gambar/Train'
test_folder = 'Gambar/Test'

#Membuat folder Train dan Test
#os.mkdir(train_folder)
#os.mkdir(test_folder)

#Membuat grup objek
cols = ['filename','id','center_x','center_y','w','h']
groupby_obj_train = train_df[cols].groupby('filename')
groupby_obj_test = test_df[cols].groupby('filename')
```

```python
#groupby_obj_train.get_group("(506).png")
#Save setiap image di train/test_folder dan save file labelnya dengan format .txt
def save_data(filename,folder_path,group_obj):
    #Move image
    src = os.path.join('Gambar',filename) #src = source
    dst = os.path.join(folder_path,filename) #dst = destination
    move(src,dst) #--> Memindahkan image di destinasi folder
    
    #Save label
    text_filename = os.path.join(folder_path, os.path.splitext(filename)[0] + '.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename,sep=' ',index=False,header=False)
```

```python
groupby_obj_train.groups.keys()
```

```python
#Train
filename_series_train = pd.Series(groupby_obj_train.groups.keys())
filename_series_train.apply(save_data,args=(train_folder,groupby_obj_train))
```

```python
#Test
filename_series_test = pd.Series(groupby_obj_test.groups.keys())
filename_series_test.apply(save_data,args=(test_folder,groupby_obj_test))
```
